```python
import math

def median_filter(pic,k=3):
  mid = math.floor(k / 2)
  pic_new = np.pad(pic,mid, mode='edge')
  a = pic_new.shape[0]
  b = pic_new.shape[1]
  new = []
  for i in range(mid,a - mid):
    row = []
    for j in range(mid,b - mid):
      dd = np.median(pic_new[i-mid:i+mid+1,j-mid:j+mid+1])
      row.append(dd)
    new.append(row)
  new = np.array(new)
  return new
```

```python
smooth = img;
for i in range(5):
  smooth = median_filter(smooth, 5)

plt.imshow(smooth, cmap='gray')
```
