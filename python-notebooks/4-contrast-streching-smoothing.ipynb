```python
import matplotlib.image as img
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = img.imread('/content/drive/MyDrive/Machine-Vision/moo2.gif')



plt.imshow(image)
```

```python
reshaped_image = image[:,:,:3]


[red, green, blue] = [reshaped_image[:,:,0], reshaped_image[:,:,1], reshaped_image[:,:,2]]

```

```python
plt.hist(red)
```

```python
input_min = np.percentile(red, 5)

input_max = np.percentile(red, 95)

print({"input_minimum" : input_min, "input_maximum" : input_max})
```

```python

input_min = int(input_min)
input_max = int(input_max)
output_min = 0
output_max = 255

intermediate = (((output_max - output_min) / (input_max - input_min)) + output_min)

def contrast_stretch(pixel):

  final = (pixel - input_min) * intermediate

  value = 0;
  if final < 0:
      value = 0;
  if final > 255:
      value = 255;

  value = int(final)

  return value



```

```python
print(red)
```

```python
new_red = [[contrast_stretch(x) for x in arr] for arr in red]


new_red = np.array(new_red)
```

```python
plt.imshow(new_red, cmap='gray')
```
